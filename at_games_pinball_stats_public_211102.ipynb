```python
#@title Base code
import getpass
import requests

import plotly.express as px
import pandas as pd
import plotly.express as px
import pandas as pd

from google.colab import data_table

from IPython.display import Image
from PIL import Image

import requests
import io
from PIL import ImageEnhance

import base64
from IPython.display import HTML

def getImage(url):
  return Image.open(io.BytesIO(requests.get(url, stream=True).content))


def downloadTopImages(scores):
  df = pd.DataFrame(data = scores)
  df.sort_values('rank', ascending = True, inplace = True) # high to low
  df = df.head(9)
  df.sort_values('rank', ascending = False, inplace = True) # then low to high

  seriesArray = df['boxart'].array
  imageArray = []
  for image in seriesArray:
    print(f'Downloading: {image}')
    imageArray.append(getImage(image))
  return imageArray


def generateAndDisplayBadge(imageArray):
  image1_size = imageArray[0].size
  halfSize = (int(image1_size[0]/2), int(image1_size[1]/2))

  new_image = Image.new('RGB',(3*image1_size[0], 3*image1_size[1]), (250,250,250))
  new_image.paste(imageArray[0])
  new_image.paste(imageArray[1], (image1_size[0]*2, 0))
  new_image.paste(imageArray[2], (image1_size[0], 0))

  new_image.paste(imageArray[3], (0, image1_size[1]))
  new_image.paste(imageArray[4], (image1_size[0]*2, image1_size[1]))

  new_image.paste(imageArray[5], (0, image1_size[1]*2))
  new_image.paste(imageArray[6], (image1_size[0]*2, image1_size[1]*2))
  new_image.paste(imageArray[7], (image1_size[0], image1_size[1]*2))

  enhancer = ImageEnhance.Brightness(new_image)

  new_image = enhancer.enhance(.75)
  bigImage = imageArray[8].resize((int(image1_size[0]*1.5),int(image1_size[1]*1.5)))
  bigSpot = (new_image.size[0]/2, new_image.size[1]/2)
  bigSpot = (bigSpot[0] - (bigImage.size[0]/2), bigSpot[1] - (bigImage.size[1]/2))
  bigSpot = (int(bigSpot[0]), int(bigSpot[1]))
  new_image.paste(imageArray[8], (image1_size[0], image1_size[1]))
  new_image.paste(bigImage, bigSpot)

  display(new_image)

  img_byte_arr = io.BytesIO()
  new_image.save(img_byte_arr, format='PNG')
  img_byte_arr = img_byte_arr.getvalue()

  b64 = base64.b64encode(img_byte_arr)
  payload = b64.decode()
  html = '<a download="tablesbadge.png" href="data:image/png;base64,{payload}" target="_blank">Download your tables badge</a>'
  html = html.format(payload=payload)
  display(HTML(html))

def requestCredentials():
  username = input('Enter your Arcade.Net e-mail: ')
  password = getpass.getpass('Enter your password: ')
  return {'username':username, 'password':password}

# TODO: Why did I have this decalred 2x?
# def login(username, password):
#   payload = {'email': username, 'password': password}
#   headers = {
#             "accept": "application/json, text/plain, */*",
#             "accept-language": "en-US,en;q=0.9",
#             "content-type": "application/json",
#             "fp": "ZDRlMGZkNmVhYmE2NTczNTNhMzQ3ZmFiOTc2NWNiMWU=",
#             "sec-ch-ua": "\" Not A;Brand\";v=\"99\", \"Chromium\";v=\"90\", \"Google Chrome\";v=\"90\"",
#             "sec-ch-ua-mobile": "?0",
#             "sec-fetch-dest": "empty",
#             "sec-fetch-mode": "cors",
#             "sec-fetch-site": "same-origin"
#         }
#   r = requests.post('https://www.atgames.net/arcadenet/backend/api/account/sign_in',json = payload, headers = headers, verify = False)

def login(username, password):
  payload = {'email': username, 'password': password}
  headers = {
            "accept": "application/json, text/plain, */*",
            "accept-language": "en-US,en;q=0.9",
            "content-type": "application/json",
            "fp": "ZDRlMGZkNmVhYmE2NTczNTNhMzQ3ZmFiOTc2NWNiMWU=",
            "sec-ch-ua": "\" Not A;Brand\";v=\"99\", \"Chromium\";v=\"90\", \"Google Chrome\";v=\"90\"",
            "sec-ch-ua-mobile": "?0",
            "sec-fetch-dest": "empty",
            "sec-fetch-mode": "cors",
            "sec-fetch-site": "same-origin"
        }
  r = requests.post('https://www.atgames.net/arcadenet/backend/api/account/sign_in',json = payload, headers = headers, verify = False)


  token = r.json()['account']['token']
  return f'Bearer {token}'

maxSize = 5
postUrl = "&after="

def simpleFetch(token, url):
  # print (f'started {url}')

  headers = {
      'Authorization': token
  }
  r = requests.get(url, headers = headers, verify = False) # Fix this
  # print(r.json())

  return r.json()

def getAllScores():
  scoresUrl = 'https://www.atgames.net/arcadenet/backend/d2d/arcade/v2/leaderboards/personal?limit=5&model=HA8819'
  return fetchRepeating(scoresUrl, postUrl, maxSize, bearerToken)

def fetchRepeating(url, postUrl, maxSize, bearerToken):
  all = []

  new = simpleFetch(bearerToken, url)
  all = all + new

  while(len(new) >= maxSize):
    gameId = new[len(new) - 1].get('game_id')
    newUrl = f'{url}{postUrl}{gameId}'
    new = simpleFetch(bearerToken, newUrl)
    all = all + new
    print(f'Fetched {len(new)} for a total {len(all)} from {url}')
  return all

def getAllGames():
  allGamesUrl = 'https://www.atgames.net/leaderboards/leaderboard/after?series=3'
  return fetchRepeating(allGamesUrl, postUrl, maxSize, bearerToken)

%load_ext google.colab.data_table
import pandas as pd

def isNotAvailable(game):
  # Want to filter stuff that is streaming, right now we have a handful of streaming and pinball tables that I don't want to see
  # Will probably need to tweak this in the future
  hardware = game["hardware_models"]
  length = len(hardware)
  return length == 2 or length == 3 or len(game["rankings"]) == 0

def isAvailable(game):
  return not isNotAvailable(game)

def displayUnplayed(allGames, allScores):
  df1 = pd.DataFrame(data = allScores)
  df2 = pd.DataFrame(data = allGames)

  df = df2[~(df2['name'].isin(df1['name']))].reset_index(drop=True) # showing df2, if it's in df2 but not df1 - this shows the unplayed games, but we need ot remove unknown/streaming games
  df = df[df.apply(isAvailable, axis=1)]
  df.drop(labels=df.columns.difference(['name']), axis = 1, inplace = True)
  # df.columns = ['Name']
  renameColumns(df)
  df.set_index('Name', inplace=True)

  display(df)

def displayUnavailableGames(allGames):
  # difference.filter(inplace = True)
  df = pd.DataFrame(data = allGames)
  df = df[df.apply(isNotAvailable, axis=1)]
  df.drop(labels=df.columns.difference(['name']), axis = 1, inplace = True)
  # df.columns = ['Name']
  renameColumns(df)
  df.set_index('Name', inplace=True)

  display(df)

def dfLimitedColumnsInPlace(df):
  return df.drop(labels=df.columns.difference(['rank','score','name']), axis = 1, inplace = True)

def renameColumns(df):
  # df.columns = ['Name','Score','Rank']
  df.rename(lambda x: x.capitalize(), axis = 1, inplace = True)

def showOnlyTop10(df):
  return df[df['rank'] <= 10]

def displayTop10s(scores):
  df = pd.DataFrame (data = scores)

  dfLimitedColumnsInPlace(df)
  df = showOnlyTop10(df)
  df.sort_values('rank', ascending = True, inplace = True)

  df['score'] = df['score'].astype(float)

  renameColumns(df)
  df.set_index('Rank', inplace=True)

  display(df)

def display10Worst(scores):
  df = pd.DataFrame(data = scores)
  dfLimitedColumnsInPlace(df)
  df.sort_values('rank', ascending = False, inplace = True)

  renameColumns(df)
  df.set_index('Rank', inplace=True)

  display(df.head(10))

def displayAllScores(scores):
  df = pd.DataFrame(data = scores)
  dfLimitedColumnsInPlace(df)
  df.sort_values('rank', ascending = True, inplace = True)
  renameColumns(df)
  df.set_index('Rank', inplace=True)

  display(df)

def displayCalcs():
  GAME_COUNT = 137

  # if (len(allScores) == GAME_COUNT):
  df = pd.DataFrame(data = allScores)

  # df['intRank'] = df['rank'].apply(lambda x: pd.to_numeric(x, errors='coerce'))
  print(f"Total Rank (Lower is better): {df['rank'].sum()}")
  # print(f"Mean: {df['rank'].mean()}")
  # print(f"Median: {df['rank'].median()}")
  print(df['rank'].describe())
  # else:
  #   print('You haven\'t placed on all of the tables yet.')

def showScatterRank(scores):
  # count for each rank
  data = dict()

  # convert to counts
  for score in allScores:
    rank = str(score['rank'])
    rankDict = data.get(rank, dict())
    rankDict['rank'] = rank
    rankDict['size'] = rankDict.get('size', 0) + 1

    namesArray = rankDict.get('names', [])
    namesArray.append(score['name'])
    rankDict['names'] = namesArray

    data[rank] = rankDict

  df = pd.DataFrame(data = data.values())
  df
  display(px.scatter(df, x = "rank", y = "rank", color = "rank", size='size', hover_data=['names'],
            labels = {'rank':'Rank', 'names':'names','size':'size'}))

def showScatterScores(scores):
  # count for each rank
  data = []

  for score in scores:
    score['score'] = float(score['score'])
    data.append(score)

  df = pd.DataFrame(data = data)
  display(px.scatter(df, x = "internal_number", y = "rank", color = "rank", size = 'score', hover_data=['name','score']))

from IPython.display import display, Markdown, clear_output

def doAllOutput():
  # clear_output(True)
  # Googles interactive tables don't honor dataframe formatters, so we need to use this
  # strangely, doesnt always format correctly, a second run seems ok though?

  display(Markdown('# 10 Worst Ranks'))
  display10Worst(allScores)

  display(Markdown('# Top 10s'))
  displayTop10s(allScores)

  display(Markdown('# Games w/o High Scores'))
  displayUnplayed(allGames, allScores)

  display(Markdown('# All Scores'))
  displayAllScores(allScores)

  display(Markdown('# Unavailable/Streaming Games'))
  displayUnavailableGames(allGames)

  display(Markdown('# Misc Stats'))
  displayCalcs()

  display(Markdown('# Charts'))
  display(Markdown('Size is Score'))
  showScatterScores(allScores)

  display(Markdown('Size is count of ranks'))
  showScatterRank(allScores)
  # data_table._DEFAULT_FORMATTERS[float] = lambda x: f"{x:,.0f}"

  # right of right is deluxe
  # left is gottlieb
  # middle of right is retro

data_table._DEFAULT_FORMATTERS[float] = lambda x: f"{x:,.0f}"
```
